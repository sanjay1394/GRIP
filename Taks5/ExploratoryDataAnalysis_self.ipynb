# Task 5 : To explore Business Analytics

## Importing important libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading data from .csv file

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

## Checking the size of the data and any missing data and unique data elements

```python
print(f'Number of rows and columns: {df.shape}')
print(f'Missing data if any: \n{df.isnull().sum()}')
print(f'Unique data numbers: \n{df.nunique()}')
```

## Checking what information the columns are displaying and their ranges

```python
df.info()
df.describe()
```

## Analysing the corelations among the columns and visualising it

```python
cr = (df.corr())
print(cr)
plt.figure(figsize=(12,6))
sns.heatmap(cr, annot = True)
plt.show()
```

## 'Ship mode' data visualisation

```python
print (df['Ship Mode'].value_counts())
plt.figure(figsize=(10,6))
sns.countplot(x=df['Ship Mode'])
plt.show()
```

## 'Segment' data visualisation

```python
print (df['Segment'].value_counts())
plt.figure(figsize=(10,6))
sns.countplot(x=df['Segment'])
plt.show()
```

## 'City' data visualisation

```python
print (df['City'].value_counts())
plt.figure(figsize=(15,6))
sns.countplot(x=df['City'], order=(df['City'].value_counts().head(50)).index)
plt.xticks(rotation=90)
plt.show()
```

## 'State' data visualisation

```python
print (df['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

## 'Region' data visualisation

```python
print (df['Region'].value_counts())
plt.figure(figsize=(10,6))
sns.countplot(x=df['Region'])
plt.show()
```

## 'Category' data visualisation

```python
print (df['Category'].value_counts())
plt.figure(figsize=(10,6))
sns.countplot(x=df['Category'])
plt.show()
```

## 'Sub-category' data visualisation

```python
print (df['Sub-Category'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
plt.figure(figsize=(12,6))
plt.pie(df['Sub-Category'].value_counts(), labels=df['Sub-Category'].value_counts().index, startangle=180, radius=1)
plt.show()
```

## 'Sub-category' with respect to 'profits' visualisation

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df['Sub-Category'], y=df['Profit'])
plt.xticks(rotation=90)
plt.show()
```
